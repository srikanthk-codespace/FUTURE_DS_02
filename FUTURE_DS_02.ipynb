```python
import pandas as pd

data = {
    'Stage': ['Visitors', 'Leads', 'Opportunities', 'Customers'],
    'Count': [1000, 400, 200, 80]
}

df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(df['Stage'], df['Count'])
plt.title("Marketing Funnel Analysis")
plt.xlabel("Stages")
plt.ylabel("Number of Users")
plt.show()
```

```python
df['Conversion Rate (%)'] = df['Count'].pct_change() * 100
df
```

```python
print("Biggest drop is from Visitors to Leads.")
print("Conversion improves in later stages.")
print("Final customer count is low compared to initial visitors.")

print("\nConclusion:")
print("Business should improve lead generation and reduce early-stage drop-offs.")
```
